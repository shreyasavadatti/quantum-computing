```python

from qiskit import QuantumCircuit
qc=QuantumCircuit(3,3)
qc.measure([0,1,2],[0,1,2])
qc.draw()
```

```python
from qiskit.providers.aer import AerSimulator
sim=AerSimulator()
job=sim.run(qc)
result=job.result()
result.get_counts()
```

```python
qc = QuantumCircuit(3, 3)
qc.x([0,1])
qc.measure([0,1,2],[0,1,2])
qc.draw()
```

```python
job = sim.run(qc)     
result = job.result()  
result.get_counts()   
```

```python
qc = QuantumCircuit(2, 2)
qc.x(0)
qc.cx(0,1)  # CNOT controlled by qubit 0 and targeting qubit 1
qc.measure([0,1], [0,1])
display(qc.draw())     

job = sim.run(qc)      
result = job.result() 
print("Result: ", result.get_counts())
```
